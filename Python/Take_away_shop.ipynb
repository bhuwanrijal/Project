```python
# declaring global variable 
shift_pizza_price=[]
shift_pasta_price=[]
large_pizza=10.99
large_pasta=9.5
choice='y'
```

```python
# dislpay the menu item with promotion 
print("\t\t\t\t The pizzera take way shop")
print("------------------------------------------------------------------------------------- \n")
print("The price list of pizza and pasta ")
print("1 large pizza.............................................. 10.99 AUD")
print("2 large pizza.............................................. 20.99 AUD")
print("3 large pizza.............................................. 29.99 AUD")
print("1 large pasta................................................ 9.5 AUD")
print("2 large pasta.............................................. 17.00 AUD")
print("3 large pasta.............................................. 27.50 AUD \n")
print("The menu item with promotion ")
print("Every 4 large pizza.......................... 1 free 1.5lt soft drink")
print("Every 4 large pasta....................... 1 free plate of bruschetta")
print("Every 4 pizzas and pastas............. 2 free choco brownie ice cream \n")
```

```python
#while loop is used to take more order in a shift 
while choice=='y':
#get input from the user to store the number of pizza
    while True:
        disp=input("Enter the number of large pizza ")
        val=int(disp)
        if val<0:
            print("Enter only positive value")
            continue
        else:
            break
# defining a function to calculate the pizza price 
    def num_pizza():
        if(val==1):
            return large_pizza
        elif(val==2):
            two_pizza_price=20.99
            return two_pizza_price
        elif(val==3):
            three_pizza_price=29.99
            return three_pizza_price
        else:
            return val*large_pizza
#get input from the user to store the number of pasta
    while True:
        disp1=input("Enter the number of large pasta ")
        val1=int(disp1)
        if val1<0:
            print("Enter only positive value")
            continue
        else:
            break
# declaring a function to calculate the pasta price 
    def num_pasta():
        if(val1==1):
            return large_pasta
        elif(val1==2):
            two_pasta_price=17.00
            return two_pasta_price
        elif(val1==3):
            three_pasta_price=27.50
            return three_pasta_price
        else:
            return val1*large_pasta
# declaring a function to check number of pizza is more than 4 
    def pizza_offer():
        if (val>=4):
            no_softdrinks=1
            return no_softdrinks
        else:
            return 0
# declaring a function to check number of pasta is more than 4
    def pasta_offer():
        if (val1>=4):
            no_bruschetta=1
            return no_bruschetta
        else:
            return 0
# declaring a function to check number of pizza and pasta is more than 4
    def pizza_pasta_offer():
        if (val>=4 & val1>=4):
            no_icecream=2
            return no_icecream
        else:
            return 0
    print("*****************************Invoice*************************************")
    total=num_pizza()+num_pasta()
    #append funtion is used to add number of pizza in shift_pizza_price
    shift_pizza_price.append(num_pizza())
    shift_pasta_price.append(num_pasta())
    #len method is used to count the number of pizza and pasta
    pizza=len(shift_pizza_price)
    pasta=len(shift_pasta_price)
    num_pizza_pasta=pizza+pasta
    #sum method is used to add the value
    sum_pizza=sum(shift_pizza_price)
    sum_pasta=sum(shift_pasta_price)
    grand_total=sum_pizza+sum_pasta
    no_pizza=pizza_offer()
    no_pasta=pasta_offer()
    no_pizza_pasta=pizza_pasta_offer()
    print("Total payment of each order is...................................." +str(total))
    print ("Total payment  each pizza order.................................." +str(num_pizza()))
    print("payment for each pasta order......................................" +str(num_pasta()))
    print ("payment recived for  pizza order................................." +str(sum_pizza))
    print("Total payment recieved for  pasta order..........................." +str(sum_pasta))
    print("Grand total payment recieved for pizza and pasta.................." +str(grand_total))
    print("Total number of pizza............................................." +str(pizza)) 
    print("Total number of pasta............................................." +str(pasta)) 
    print("Grand total number of pizza and pasta............................." +str(num_pizza_pasta))    
    print("Total number of 1.5lt bottle of soft drinks......................." +str(no_pizza))
    print("Total number of bruschetta plates................................." +str(no_pasta))
    print("Total number of Choco brownie ice cream..........................." +str(no_pizza_pasta))
    tot=input("press y to continue or press any key to exit \n")
    if tot=='y':
        continue
    else:
        print("Total payment recived for  pizza order in a shift................................." +str(sum_pizza))
        print("Total payment recieved for  pasta order in a shift..........................." +str(sum_pasta))
        print("Total number of pizza in a shift............................................." +str(pizza)) 
        print("Total number of pasta in a shift............................................." +str(pasta))
        break
   
```
